### LorentzVector class

```scala
type CalculableType = Float;
val _pi: Float = 3.1415F

def Phi_mpi_pi(xxx: CalculableType) = {
    var tmp = xxx
    while (tmp >= _pi) 
        tmp -= 2 * _pi
    while (tmp < - _pi)
        tmp += 2 * _pi
    tmp
}

case class LorentzVector(x: CalculableType = 0, y: CalculableType = 0, z: CalculableType = 0, t: CalculableType = 0) {
    def SetPtEtaPhiM(ptt: CalculableType, eta: CalculableType, phi: CalculableType, mass: CalculableType): LorentzVector = {
        val pt: CalculableType = math.abs(ptt)
        this.SetXYZM(pt*math.cos(phi).toFloat, pt*math.sin(phi).toFloat, pt*math.sinh(eta).toFloat, mass)
    }
        
    def SetXYZM(x: CalculableType, y: CalculableType, z:CalculableType, m: CalculableType): LorentzVector = {
        val t: Float = 
            if (m >= 0) 
                math.sqrt(x*x + y*y + z*z + m*m).toFloat
            else 
                math.sqrt(math.max(x*x + y*y + z*z - m*m, 0)).toFloat
        LorentzVector(x,y,z,t)
    }
            
    def E:CalculableType = t
    
    def Px:CalculableType = x
    
    def Py:CalculableType = y
    
    def Pz:CalculableType = z
    
    def Pt:CalculableType = math.sqrt(x*x + y*y).toFloat
    
    def Eta: CalculableType = {
        val cosTheta = this.CosTheta
        if (cosTheta*cosTheta<1)
            return -0.5F*math.log((1.0 - cosTheta)/(1.0 + cosTheta)).toFloat
        if (z == 0) 
            return 0.0F
        return 0.0F
    }
    
    def mag: CalculableType = math.sqrt(x*x + y*y + z*z).toFloat
    
    def CosTheta:CalculableType = if (this.mag==0.0) 1.0F else z/this.mag
    
    def Phi:CalculableType = math.tan(y/x).toFloat
    
    def DeltaR(other: LorentzVector):CalculableType = {
        val deta = this.Eta - other.Eta
        val dphi = Phi_mpi_pi(this.Phi - other.Phi)
        math.sqrt(deta*deta + dphi*dphi).toFloat
    }
}
    
def mysign_func(v: CalculableType): Int = 
    if (v<0) return -1
    else if (v==0) return 0
    else return 1
```

```scala
// works fine for spark 2.3 not for 2.2 - does not like the aliases!!!
// TODO: Get the alias in!
type FeatureType = Seq[Seq[Float]]

def ChPtMapp(dr: Double, allTracks: Seq[Record0]): Seq[Seq[Float]] = 
    allTracks.filter(_.PT > 0.5).map({case track => Seq(track.Eta, track.Phi, track.PT)})

def NeuPtMapp(dr: Double, allNeutrals: Seq[Record3]): Seq[Seq[Float]] = 
    allNeutrals.filter(_.ET > 1.0).map({case neutral => Seq(neutral.Eta, neutral.Phi, neutral.ET)})

def PhotonPtMapp(dr: Double, allPhotons: Seq[Record4]): Seq[Seq[Float]] = 
    allPhotons.filter(_.ET > 1.0).map({case photon => Seq(photon.Eta, photon.Phi, photon.ET)})

def PFIso(p4: LorentzVector, dr: Float, selected: Seq[Seq[Float]], shouldSubtractPt: Boolean): Float = {
    if (p4.Pt <= 0.0) return 0.0F
    val sumPtsWithinCone: Float = selected.map({p: Seq[Float] => 
        (((p(0)-p4.Eta)*(p(0) - p4.Eta) + (Phi_mpi_pi(p(1) - p4.Phi)*Phi_mpi_pi(p(1) - p4.Phi))) < dr*dr, p(2))
    }).filter({case pair => pair._1 == true}).map({case pair => pair._2}).fold[Float](0.0F)(_ + _)
    val iso: Float = (sumPtsWithinCone/p4.Pt).toFloat
    if (shouldSubtractPt) iso - 1 else iso
}
```

```scala
def buildLeptonFeatures(electrons: Seq[Record5], 
                  muons: Seq[Record8], 
                  selectedTracks: Seq[Seq[Float]], 
                  selectedNeutrals: Seq[Seq[Float]], 
                  selectedPhotons: Seq[Seq[Float]]): Seq[Float] = {
    if (electrons.size == 0 && muons.size==0) return Seq.empty[Float]
    var foundMuon: Option[Seq[Float]] = None
    var foundElectron: Option[Seq[Float]] = None
    
    //
    // Select valid electrons and pick the first one
    //
    val validEles: Seq[Seq[Float]] = (for (ele <- electrons if ele.PT > 25;
         l = LorentzVector().SetPtEtaPhiM(ele.PT, ele.Eta, ele.Phi, 0);
         isoCh = PFIso(l, 0.3F, selectedTracks, true);
         isoNeu = PFIso(l, 0.3F, selectedNeutrals, false);
         isoGamma = PFIso(l, 0.3F, selectedPhotons, false);
         if (isoCh + isoNeu + isoGamma) < 0.2F) yield Seq(l.E, l.Px, l.Py, l.Pz, 
                                                         l.Pt, l.Eta, l.Phi, 0, 0, 0,
                                                         isoCh, isoGamma, isoNeu, 0, 0,
                                                         0, 1, 0, ele.Charge)).asInstanceOf[Seq[Seq[Float]]]
    foundElectron = if (validEles.size == 0) None else Some(validEles.head)
  
    //
    // Select valid muons and pick the first one
    //
    val validMuons: Seq[Seq[Float]] = (for (muon <- muons if muon.PT > 25F;
                          l = LorentzVector().SetPtEtaPhiM(muon.PT, muon.Eta, muon.Phi, 0);
                          isoCh = PFIso(l, 0.3F, selectedTracks, true);
                          isoNeu = PFIso(l, 0.3F, selectedNeutrals, false);
                          isoGamma = PFIso(l, 0.3F, selectedPhotons, false); 
                          result = Seq(l.E, l.Px, l.Py, l.Pz,
                                       l.Pt, l.Eta, l.Phi, 0F, 0F, 0F, 
                                       isoCh, isoGamma, isoNeu,
                                       0F, 0F, 0F, 0F, 1F, muon.Charge.toFloat)
                          if (isoCh + isoNeu + isoGamma) < 0.2F) yield result).asInstanceOf[Seq[Seq[Float]]]
    foundMuon = if (validMuons.size == 0) None else Some(validMuons.head)
    
    //
    // unwrap options
    //
    (foundElectron, foundMuon) match {
        case (Some(e), Some(m)) => if (e(5) > m(5)) e else m
        case (Some(e), None) => e
        case (None, Some(m)) => m
        case (None, None) => Seq()
        case (_, _) => Seq()
    }
}
```

```scala
case class FeatureSet(leptonFeatures: Seq[CalculableType] = Nil, 
                      trackFeatures: Seq[Seq[Float]] = Nil, 
                      neutralFeatures: Seq[Seq[Float]] = Nil,
                      photonFeatures: Seq[Seq[Float]] = Nil, isEmpty: Boolean = true)

def buildFeatures(e: Event): FeatureSet = {
    // select tracks
    val selectedTracks = ChPtMapp(0.3, e.EFlowTrack)
    if (selectedTracks.size == 0) return FeatureSet()
    
    // select neutrals
    val selectedNeutrals = NeuPtMapp(0.3, e.EFlowNeutralHadron)
    if (selectedNeutrals.size == 0) return FeatureSet()
    
    // select photons
    val selectedPhotons = PhotonPtMapp(0.3, e.EFlowPhoton)
    if (selectedPhotons.size == 0) return FeatureSet()
    
    // lepton features
    val leptonFeatures: Seq[Float] = buildLeptonFeatures(e.Electron, e.MuonTight, 
                                             selectedTracks, selectedNeutrals, selectedPhotons)
    if (leptonFeatures.size==0) return FeatureSet()
    // lepton is good to go
    val p4Lepton = LorentzVector(leptonFeatures(1), leptonFeatures(2), leptonFeatures(3), leptonFeatures(0))
    
    // build Track Features
    val trackFeatures = (for (track <- e.EFlowTrack if track.PT > 0.5F;
                              p4 = LorentzVector().SetPtEtaPhiM(track.PT, track.Eta, track.Phi, 0.0F) if p4Lepton.DeltaR(p4) > 0.0001;
                              isoCh = PFIso(p4, 0.3F, selectedTracks, true);
                              isoNeu = PFIso(p4, 0.3F, selectedNeutrals, false);
                              isoGamma = PFIso(p4, 0.3F, selectedPhotons, false)) yield Seq(p4.E, p4.Px, p4.Py, p4.Pz, 
                                                                      track.PT, track.Eta, track.Phi, track.X,
                                                                      track.Y, track.Z, isoCh, isoGamma, isoNeu,
                                                                      1F, 0F, 0F, 0F, 0F, mysign_func(track.PID).toFloat)).asInstanceOf[Seq[Seq[Float]]]
    
    // build photon features
    val photonFeatures = (for (photon <- e.EFlowPhoton if photon.ET > 1.0F;
                                p4 = LorentzVector().SetPtEtaPhiM(photon.ET, photon.Eta, photon.Phi, 0.0F);
                                isoCh = PFIso(p4, 0.3F, selectedTracks, true);
                                isoNeu = PFIso(p4, 0.3F, selectedNeutrals, false);
                                isoGamma = PFIso(p4, 0.3F, selectedPhotons, false)) yield Seq(p4.E, p4.Px, p4.Py, p4.Pz,
                                                                                             photon.ET, photon.Eta, photon.Phi,
                                                                                             0F, 0F, 0F, isoCh, isoGamma, isoNeu,
                                                                                             0F, 0F, 1F, 0F, 0F, 0F)).asInstanceOf[Seq[Seq[Float]]]
    
    // build neutral hadron features
    val neutralFeatures = (for (neu <- e.EFlowNeutralHadron if neu.ET > 1.0;
                                p4 = LorentzVector(neu.ET, neu.Eta, neu.Phi, 0.0F);
                                isoCh = PFIso(p4, 0.3F, selectedTracks, true);
                                isoNeu = PFIso(p4, 0.3F, selectedNeutrals, false);
                                isoGamma = PFIso(p4, 0.3F, selectedPhotons, false)) yield Seq(p4.E, p4.Px, p4.Py, p4.Pz, 
                                                                                             neu.ET, neu.Eta, neu.Phi, 0F, 0F, 0F,
                                                                                             isoCh, isoNeu, isoGamma, 
                                                                                             0F, 1F, 0F, 0F, 0F, 0F)).asInstanceOf[Seq[Seq[Float]]]
    FeatureSet(leptonFeatures, trackFeatures, neutralFeatures, photonFeatures, false)
}
```

```scala
val before = System.nanoTime
val features = events
    .limit(5000)
    .map({e: Event => buildFeatures(e)})
    .toDF("leptonFeatures", "trackFeatures", "neutralFeatures", "photonFeatures", "isEmpty")
    .filter("isEmpty == false")
val after = System.nanoTime
val exeTime = (after - before)/1000000000
println(s"Total Execution Time $exeTime s")
```
